```python
import pandas as pd
```

```python
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')
```

```python
## Q1 Dataset Exploration
df.head(10)
df.info()
df.shape
```

```python
## Q2 Sales Filtering Analysis
df[df['Sales']>3500]
df[(df['Region'] == "East") & (df['Sales'] > 2500)]
```

```python
## Q3 Customer Purchase Analysis
df.groupby('Customer')['Sales'].agg('sum').sort_values(ascending =False).head()
```

```python
## Q4 Regional Sales Performance
df.groupby('Region')['Sales'].agg('sum').sort_values(ascending = False).head(1)
```

```python
## Q5 Product Performance Analysis
df.groupby('Product')['Quantity'].sum()
df.groupby('Product')['Quantity'].sum().sort_values(ascending=False).head(1) 
```

```python
## Q6 Multi-Level Grouping
df.groupby(['Region','Category'])['Sales'].agg('sum')
df.groupby(['Region','Customer','City','Product'])['Sales'].agg('sum').sort_values(ascending=False)
## Through this analysis it can be said that the most sales is coming from the 'East' region from City 'Ahmedabad' and from the phone in electronics category which is giving the highest sale.
```

```python
## Q7 Pivot Table Creation
pivot = df.pivot_table(values='Sales', columns ='Product', index='Region', aggfunc ='sum')
print(pivot)
```

```python
## Q8 Profitability Analysis
df.groupby(['Category','Product'])['Profit'].mean()
df.groupby('Category')['Profit'].sum().sort_values(ascending = False).head(1)
```

```python
## Q9 Discount Impact Study
df.groupby("Discount")["Profit"].mean()
df.groupby("Discount")["Profit"].sum()
```

```python
## Q10 End-to-End Data Wrangling Task
df[df['Sales'] > 3000]
df.groupby(['Region','Category'])['Sales'].agg('sum')
df['Sales'].sum()
pivot = df.pivot_table(values='Sales', columns ='Category', index='Region', aggfunc ='sum')
print(pivot)
```
